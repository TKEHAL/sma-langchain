# LangChain agent with Groq

This notebook uses the LangChain v1 agent API. The important part is that `agent.invoke(...)` receives a dictionary with a `messages` list.

```python
from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain_groq import ChatGroq

load_dotenv(override=True)

GROQ_MODEL = "llama-3.1-8b-instant"
assert not GROQ_MODEL.startswith("groq:"), "Remove the 'groq:' prefix when using ChatGroq."

llm = ChatGroq(
    model=GROQ_MODEL,
    temperature=0,
)

print("Using Groq model:", llm.model_name)

agent = create_agent(
    model=llm,
    tools=[],
    system_prompt="You are a helpful assistant.",
)

response = agent.invoke(input={
    "messages": [
        {"role": "user", "content": "Say hello in one short sentence."}
    ]
})

response["messages"][-1].content
```

## Reuse the agent

Run the first code cell before this one. If you changed the model or had an earlier error, restart the kernel and run all cells so the `agent` variable is recreated.

```python
from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain_groq import ChatGroq

load_dotenv(override=True)

GROQ_MODEL = "llama-3.1-8b-instant"
assert not GROQ_MODEL.startswith("groq:"), "Remove the 'groq:' prefix when using ChatGroq."

llm = ChatGroq(model=GROQ_MODEL, temperature=0)
print("Using Groq model:", llm.model_name)

agent = create_agent(
    model=llm,
    tools=[],
    system_prompt="You are a helpful assistant.",
)

question = "Je m'appelle Taoufik. Reponds-moi en francais."

result = agent.invoke(input={
    "messages": [
        {"role": "user", "content": question}
    ]
})

print(result["messages"][-1].content)
```
